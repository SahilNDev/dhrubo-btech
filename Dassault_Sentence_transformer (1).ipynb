```python
import pandas as pd
import numpy as np
from sentence_transformers import SentenceTransformer
```

```python
# 1. Load the Excel file
df_cleaned = pd.read_excel('/content/Data_Dassault_Cleaned.xlsx')
```

```python
# 2. Load a pretrained Sentence Transformer model
model = SentenceTransformer("all-MiniLM-L6-v2")
```

```python
# 3. Define your query
query = "firearms less than .50 caliber"

# 4. Calculate embeddings for the query and the descriptions
embedding1 = model.encode(query)
embeddings = model.encode(df_cleaned["Description"])
# 5. Calculate the embedding similarities
similarities = np.dot(embeddings, embedding1)  # Dot product similarity
print(list(similarities))

# 6. Find the best match and retrieve the corresponding 'Full_Code'
best_match_idx = np.argmax(similarities)
best_match_code = f"USML.{df_cleaned.iloc[best_match_idx]['Full_Code']}"
print(f"The best match is: {best_match_code}")
```
